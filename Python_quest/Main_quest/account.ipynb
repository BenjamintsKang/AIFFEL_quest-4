```python
import random as r
from datetime import datetime
#계좌 ID 랜덤 생성 메서드
def generate_account_id():
    part1 = r.randint(100, 999)
    part2 = r.randint(10, 99)
    part3 = r.randint(100000, 999999)

    account_id = f"{part1:03d}-{part2:02d}-{part3:06d}"
    return account_id

# 숫자 3자리 수 마다 ',' 붙여주는 메서드
def balance_formatter(number):
  return f"{number:,}"
```

```python
# Account class 생성
class Account:
  account_num = 0
  def __init__(self, owner, balance):
    self.bank_name = 'SC은행'
    self.owner = owner
    self.account_id = generate_account_id()
    self.balance = int(balance)
    self.deposit_count = 0
    self.history = []
    Account.account_num += 1

  @classmethod # 생성된 계좌 수
  def get_account_num(cls):
    print(f"현재 생성 계좌 수: {cls.account_num}")

  # 입금
  def deposit(self):
    while True:
      try:
        print(f"현재 잔고: {self.balance}")
        value = int(input('입금할 금액을 입력하세요: '))
        if value <= 0:
          raise Exception('1원 이상 입금이 가능합니다.')
      except ValueError:
        print('잘못된 입력입니다.')
      except Exception as e:
        print(e)
      else:
        now = datetime.now()
        self.deposit_count += 1
        if self.deposit_count % 5 == 0:
          print(f"입금 {self.deposit_count}회, 현재 잔액 {balance_formatter(self.balance)}의 1%({self.balance / 100}) 이자 추가")
          self.balance += self.balance / 100
          self.history.append({'상태': "입금", '금액': self.balance / 100, '잔고': self.balance})
        self.balance += value
        self.history.append({'상태': "입금", '금액': value, '잔고': self.balance, '날짜': now.strftime('%Y-%m-%d %H:%M:%S')})
        print(f"현재 잔액: {self.balance}")
        while True:
          try:
            deposit_or_exit = int(input("계속 입금('1'), 입금 종료('2')"))
            if deposit_or_exit != 1 and deposit_or_exit != 2:
              raise Exception()
          except:
            print("잘못된 입력입니다.")
          else:
            if deposit_or_exit == 1:
              break
            else:
              return f"현재 잔액: {self.balance}"

  # 출금
  def withdraw(self):
    while True:
      try:
        print(f"현재 잔고: {self.balance}")
        value = int(input('출금할 금액을 입력하세요: '))
        if value <= 0:
          raise Exception('1원 이상 출금이 가능합니다.')
        if value > self.balance:
          raise Exception('잔고 이상으로 출금할 수 없습니다.')
      except ValueError:
        print('잘못된 입력입니다.')
      except Exception as e:
        print(e)
      else:
        now = datetime.now()
        self.balance -= value
        self.history.append({'상태': "출금", '금액': value, '잔고': self.balance, '날짜': now.strftime('%Y-%m-%d %H:%M:%S')})
        print(f"현재 잔액: {self.balance}")
        while True:
          try:
            deposit_or_exit = int(input("계속 출금('1'), 출금 종료('2')"))
            if deposit_or_exit != 1 and deposit_or_exit != 2:
              raise Exception()
          except:
            print("잘못된 입력입니다.")
          else:
            if deposit_or_exit == 1:
              break
            else:
              return f"현재 잔액: {self.balance}"

  # 계좌 정보 출력
  def display_info(self):
    print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_id}, 잔고: {balance_formatter(self.balance)}")

  # 입금 내역 출력
  def deposit_history(self):
    for part in self.history:
      if part['상태'] == '입금':
        print(part)

  #출금 내역 출력
  def deposit_history(self):
    for part in self.history:
      if part['상태'] == '출금':
        print(part)
```

```python
harry = Account("해리 포터", 100000)
iron_man = Account("아이언맨", 500000)
jack = Account("잭 스패로우", 75000)

instance_list = [harry, iron_man, jack]

Account.get_account_num()

for account in instance_list:
    account.display_info()

iron_man.deposit()
iron_man.withdraw()
for account in instance_list:
  if account.balance >= 1000000:
    print(f"잔고 100만원 이상 예금주: {account.owner}, 잔액: {balance_formatter(account.balance)}")

iron_man.deposit_history()
```
