```python
unique_nan_addresses = df[df['위도'].isna()]['주소'].unique()
location_dict = {}

def get_geocode(address):
    try:
        geo_loc = maps.geocode(address)
        lat = geo_loc[0]['geometry']['location']['lat']
        lng = geo_loc[0]['geometry']['location']['lng']
        return [lat,lng]
    except:
        return [None, None]

for address in unique_nan_addresses:
    location = get_geocode(address)
location_dict[address] = (location[0], location[1])

# def get_geocode(address):
#     try:
#         return geolocator.geocode(address)
#     except GeocoderTimedOut:
#         print('타임아웃')
#         return get_geocode(address)

# for address in unique_addresses:
#     location = get_geocode(address)
#     if location:
#         location_dict[address] = (location.latitude, location.longitude)
#     else:
#         print(address)
#         print('에러')
#         location_dict[address] = (None, None)
```
