```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
datapath = ['./seoul_2021.csv', 
            './seoul_2022.csv', 
            './seoul_2023.csv', 
            './busan_2021.csv', 
            './busan_2022.csv', 
            './busan_2023.csv']
df = [pd.read_csv(path) for path in datapath]
df = pd.concat(df, ignore_index=True)
```

```python
# 학습에 불필요한 컬럼 제거
df = df.drop(['NO', '계약일', '매수자', '매도자', '주택유형', '등기일자', '중개사소재지'], axis=1)
df
```

```python
df = df.replace('-', np.nan)
```

```python
df['거래금액(만원)'] = df['거래금액(만원)'].str.replace(',', '').astype(int)
```

```python
sigungus_id_dict = {}
current_id = 1

def assign_sigungus_id(sigungus):
    global current_id
    if sigungus not in sigungus_id_dict:
        sigungus_id_dict[sigungus] = current_id
        current_id += 1
    return sigungus_id_dict[sigungus]

df['시군구_ID'] = df['시군구'].apply(assign_sigungus_id)
df
```
